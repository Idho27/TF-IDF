```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re
```

```python
def casefolding(s):
    new_str = s.lower()
    
    return new_str

def removeNoise(str):
    teks = re.sub(r'[a-zA-Z0-9_]*.com', "", str)
    teks = re.sub(r'[^a-zA-Z_ ]', "", teks)
    return teks

def stemmingIndo(str):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    
    return stemmer.stem(str)

def term_freq(teks):
    tf = dict()
    words = teks.split()

    for word in words:
        if word in tf:
            tf[word] += 1
        else:
            tf[word] = 1

    return tf

def word_count(teks):
    count = 0
    words = teks.split()
    
    for word in words:
        count += 1
        
    return count

def mainTfIdf():
    f = open ('sumber.txt','r')
    text_data = f.read()
    f.close()    
    print(text_data)
    text_data = casefolding(text_data)
    text_data = removeNoise(text_data)
    text_data = stemmingIndo(text_data)    
    print(text_data)
    tf = term_freq(text_data)
    wc = word_count(text_data)
    print()
    print("total kata    = ", wc)
    c = 0
    for kata in tf:
        #Penghitungan TF menggunakan cara normalisasi dengan rumus TF = a + (1-a)*tf/tfmax
        #Sementara untuk penghitungan IDF masih dihiraukan karena dokumen yang digunakan hanya satu
        tf[kata] = 0.4 + 0.6*tf[kata]/wc
        c += 1
    s = pd.Series(tf)
    s = s.sort_values(ascending=False)
    print("total variasi = ", c)
    print()
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        print(s)
    
if __name__ == "__main__":
    mainTfIdf()
```
